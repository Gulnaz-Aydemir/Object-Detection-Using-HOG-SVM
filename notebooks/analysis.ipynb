```python

# Bu notebook, proje geliştirme aşamasındaki denemeleri içerir.
# Asıl kodlar src/ klasörü altındadır.

import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog


img = cv2.imread('../data/training_set/pos/pos_1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.resize(gray, (64, 64))

fd, hog_image = hog(gray, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True)

plt.imshow(hog_image, cmap='gray')
plt.title("HOG Görselleştirme Denemesi")
plt.show()
```
